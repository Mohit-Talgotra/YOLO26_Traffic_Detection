## 6. Preview outputs

```python
from pathlib import Path
from IPython.display import Image, Video, display

output_dir = Path('outputs')
latest_media = max(output_dir.glob('*'), key=lambda p: p.stat().st_mtime)
print('Latest output:', latest_media)
if latest_media.suffix.lower() in {'.jpg', '.jpeg', '.png', '.bmp', '.webp'}:
    display(Image(filename=str(latest_media)))
elif latest_media.suffix.lower() in {'.mp4', '.avi', '.mov', '.mkv', '.wmv', '.mpeg'}:
    display(Video(filename=str(latest_media), embed=True))
```
